### 문제 3 일원분산분석: 지점별 고객만족도 비교

```python
import pandas as pd
from scipy import stats

df = pd.read_csv("./type3_sample_csvs/branch_satisfaction.csv")

a = df[df["branch"] == "A"]["score"]
b = df[df["branch"] == "B"]["score"]
c = df[df["branch"] == "C"]["score"]

result = stats.f_oneway(a, b, c)

means = df.groupby("branch")["score"].mean()

print("지점별 평균:")
print(means)
print("F-statistic:", result.statistic)
print("p-value:", result.pvalue)
print("평균이 가장 높은 지점:", means.idxmax())
```

```python
from pathlib import Path
import pandas as pd
from scipy import stats

DATA_DIR = Path('./type3_sample_csvs')
file_path = DATA_DIR / 'branch_satisfaction.csv'

df = pd.read_csv(file_path)

print('[데이터 미리보기]')
print(df.head(), '\n')

a = df[df['branch'] == 'A']['score']
b = df[df['branch'] == 'B']['score']
c = df[df['branch'] == 'C']['score']

result = stats.f_oneway(a, b, c)
means = df.groupby('branch')['score'].mean().sort_values(ascending=False)

print('[분석 결과]')
print('지점별 평균')
print(means, '\n')
print('F-statistic:', round(result.statistic, 4))
print('p-value:', round(result.pvalue, 6))
print('평균이 가장 높은 지점:', means.index[0])

print('\n[결과 해석]')
if result.pvalue < 0.05:
    print('p-value가 0.05보다 작으므로, 지점 간 평균 고객만족도 차이가 유의하다고 해석할 수 있습니다.')
else:
    print('지점 간 평균 고객만족도 차이가 유의하다고 보기 어렵습니다.')
```

### 문제 4 이원분산분석: 광고유형과 연령대가 구매금액에 미치는 영향

```python

import pandas as pd
from scipy import stats

df = pd.read_csv("./type3_sample_csvs/before_after_worktime.csv")
before = df["before_min"]
after = df["after_min"]
diff = before - after
result = stats.ttest_rel(before, after, alternative="greater")

print("평균 차이(before - after):", diff.mean())
print("t-statistic:", result.statistic)
print("p-value:", result.pvalue)
```

```python
!pip install statsmodels

from pathlib import Path
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from statsmodels.stats.anova import anova_lm

DATA_DIR = Path('./type3_sample_csvs')
file_path = DATA_DIR / 'ad_age_sales.csv'

df = pd.read_csv(file_path)

print('[데이터 미리보기]')
print(df.head(), '\n')

model = smf.ols('sales_amount ~ C(ad_type) + C(age_group) + C(ad_type):C(age_group)', data=df).fit()
anova_table = anova_lm(model, typ=2)

print('[ANOVA 결과표]')
print(anova_table, '\n')

p_ad = anova_table.loc['C(ad_type)', 'PR(>F)']
p_age = anova_table.loc['C(age_group)', 'PR(>F)']
p_inter = anova_table.loc['C(ad_type):C(age_group)', 'PR(>F)']

print('[핵심 p-value]')
print('광고유형 p-value:', p_ad)
print('연령대 p-value:', p_age)
print('상호작용 p-value:', p_inter)

print('\n[결과 해석]')
if p_inter < 0.05:
    print('상호작용 효과가 유의하므로, 광고유형의 효과는 연령대에 따라 달라진다고 해석합니다.')
else:
    print('상호작용 효과가 유의하지 않으므로, 광고유형과 연령대의 주효과 중심으로 해석합니다.')
```
